## STEP 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nice
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print('✅ Libraries loaded successfully!')
```

## STEP 2 — Load the Dataset

```python
# Load the Titanic dataset (built into seaborn — no download needed!)
df = sns.load_dataset('titanic')

print('✅ Dataset loaded!')
print(f'Shape: {df.shape}')  # rows x columns
print(f'Rows: {df.shape[0]}, Columns: {df.shape[1]}')
```

## STEP 3 — First Look at the Data

```python
# See the first 5 rows
df.head()
```

```python
# See the last 5 rows
df.tail()
```

```python
# Column names and data types
print('Column names:')
print(df.columns.tolist())
print()
print('Data types:')
print(df.dtypes)
```

```python
# Statistical summary of numeric columns
df.describe()
```

```python
# Full info — shows non-null counts
df.info()
```

## STEP 4 — Find Missing Values
**This is one of the most important skills in interviews!**

```python
# Count missing values in each column
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0])
```

```python
# Show as percentage
missing_percent = (df.isnull().sum() / len(df)) * 100
print('Missing values as percentage:')
print(missing_percent[missing_percent > 0].round(2))
```

```python
# Visualise missing values
plt.figure(figsize=(10, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap (Yellow = Missing)', fontsize=14)
plt.show()
```

## STEP 5 — Handle Missing Values

```python
# Make a copy so we don't mess up the original
df_clean = df.copy()

# Fill missing AGE with the median age
median_age = df_clean['age'].median()
df_clean['age'].fillna(median_age, inplace=True)
print(f'✅ Filled missing ages with median: {median_age}')

# Fill missing EMBARKED with the most common value (mode)
most_common_embarked = df_clean['embarked'].mode()[0]
df_clean['embarked'].fillna(most_common_embarked, inplace=True)
print(f'✅ Filled missing embarked with: {most_common_embarked}')

# Drop the DECK column — too many missing values (over 70%!)
df_clean.drop(columns=['deck'], inplace=True)
print('✅ Dropped deck column (too many missing values)')

# Verify no more missing values
print(f'\nRemaining missing values: {df_clean.isnull().sum().sum()}')
```

## STEP 6 — Remove Duplicates

```python
# Check for duplicate rows
duplicates = df_clean.duplicated().sum()
print(f'Duplicate rows: {duplicates}')

# Remove duplicates if any
df_clean.drop_duplicates(inplace=True)
print(f'✅ Dataset shape after removing duplicates: {df_clean.shape}')
```

## STEP 7 — Explore the Data with Visualisations

```python
# How many survived vs died?
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='survived', palette='Set2')
plt.title('Survival Count (0 = Died, 1 = Survived)', fontsize=14)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

survival_rate = df_clean['survived'].mean() * 100
print(f'Overall survival rate: {survival_rate:.1f}%')
```

```python
# Survival by gender
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='sex', hue='survived', palette='Set1')
plt.title('Survival by Gender', fontsize=14)
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 4))
sns.histplot(data=df_clean, x='age', hue='survived', bins=30, kde=True)
plt.title('Age Distribution by Survival', fontsize=14)
plt.show()
```

```python
# Survival by passenger class
plt.figure(figsize=(6, 4))
sns.countplot(data=df_clean, x='pclass', hue='survived', palette='coolwarm')
plt.title('Survival by Passenger Class', fontsize=14)
plt.xlabel('Class (1=First, 2=Second, 3=Third)')
plt.show()
```

```python
# Correlation heatmap — which features are related?
plt.figure(figsize=(10, 6))
numeric_df = df_clean.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=14)
plt.show()
```

## STEP 8 — Feature Engineering
**Creating new useful columns from existing ones**

```python
# Create a new column: family size
df_clean['family_size'] = df_clean['sibsp'] + df_clean['parch'] + 1
print('✅ Created family_size column')

# Create age groups
df_clean['age_group'] = pd.cut(df_clean['age'], 
                                bins=[0, 12, 18, 35, 60, 100],
                                labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
print('✅ Created age_group column')
print(df_clean['age_group'].value_counts())
```

## STEP 9 — Prepare Data for ML
**Convert text columns to numbers — ML models only understand numbers!**

```python
# Convert sex to 0/1
df_clean['sex_encoded'] = df_clean['sex'].map({'male': 0, 'female': 1})
print('✅ Encoded sex column')

# One-hot encode embarked column
embarked_dummies = pd.get_dummies(df_clean['embarked'], prefix='embarked')
df_clean = pd.concat([df_clean, embarked_dummies], axis=1)
print('✅ One-hot encoded embarked column')

print('\nFinal columns:')
print(df_clean.columns.tolist())
```

## STEP 10 — Final Summary
**What you did in this notebook:**

```python
print('=' * 50)
print('EDA SUMMARY')
print('=' * 50)
print(f'Original dataset shape: {df.shape}')
print(f'Cleaned dataset shape:  {df_clean.shape}')
print(f'Missing values handled: Age (median fill), Embarked (mode fill), Deck (dropped)')
print(f'New features created:   family_size, age_group, sex_encoded')
print(f'Survival rate:          {df_clean["survived"].mean()*100:.1f}%')
print('=' * 50)
print('✅ Data is now clean and ready for ML modeling!')
```
